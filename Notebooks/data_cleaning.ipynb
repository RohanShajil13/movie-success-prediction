# Movie Success Prediction - Data Cleaning

This notebook performs data loading, merging, cleaning, and preprocessing for the movie success prediction project using the TMDB dataset.

Steps:
- Load datasets
- Merge movies and credits
- Clean and preprocess data
- Extract useful features (genres, cast, director)
- Handle missing values
- Create target variable (success)
- Save cleaned dataset

```python
import pandas as pd
import ast
import os

movies_path = "../Datasets/tmdb_5000_movies.csv"
credits_path = "../Datasets/tmdb_5000_credits.csv"

movies = pd.read_csv(movies_path)
credits = pd.read_csv(credits_path)

print("Movies shape:", movies.shape)
print("Credits shape:", credits.shape)

display(movies.head())
display(credits.head())
```

```python
print("Movies columns:")
print(movies.columns.tolist())

print("\nCredits columns:")
print(credits.columns.tolist())
```

```python
credits = credits.rename(columns={"movie_id": "id"})
df = movies.merge(credits, on="id", suffixes=("_movie", "_credit"))

print("Merged shape:", df.shape)
print(df.columns.tolist())
display(df.head())
```

```python
if "title_movie" in df.columns:
    df = df.rename(columns={"title_movie": "title"})
elif "original_title" in df.columns:
    df = df.rename(columns={"original_title": "title"})

if "title_credit" in df.columns:
    df = df.drop(columns=["title_credit"])

print(df.columns.tolist())
```

```python
columns_needed = [
    "id",
    "title",
    "genres",
    "keywords",
    "cast",
    "crew",
    "budget",
    "revenue",
    "runtime",
    "vote_average",
    "vote_count",
    "popularity",
    "release_date",
    "original_language"
]

df = df[columns_needed].copy()

display(df.head())
```

```python
def safe_parse(text):
    """
    Converts a string version of a list/dictionary into a Python object.
    Returns an empty list if parsing fails.
    """
    try:
        return ast.literal_eval(text)
    except:
        return []


def extract_names(text):
    """
    Extracts the 'name' values from JSON-like list columns.
    Used for genres and keywords.
    """
    items = safe_parse(text)
    names = []
    
    for item in items:
        if isinstance(item, dict) and "name" in item:
            names.append(item["name"])
    
    return names


def extract_top_cast(text, top_n=3):
    """
    Extracts the first top_n cast member names.
    """
    items = safe_parse(text)
    names = []
    
    for item in items:
        if isinstance(item, dict) and "name" in item:
            names.append(item["name"])
        
        if len(names) == top_n:
            break
    
    return names


def extract_director(text):
    """
    Extracts the director name from the crew column.
    """
    items = safe_parse(text)
    
    for item in items:
        if isinstance(item, dict) and item.get("job") == "Director":
            return item.get("name")
    
    return None
```

```python
df.loc[:, "genres"] = df["genres"].apply(extract_names)
df.loc[:, "keywords"] = df["keywords"].apply(extract_names)
df.loc[:, "cast"] = df["cast"].apply(extract_top_cast)
df.loc[:, "director"] = df["crew"].apply(extract_director)

df = df.drop(columns=["crew"])

display(df[["title", "genres", "keywords", "cast", "director"]].head())
```

```python
print("Missing values before cleaning:")
print(df.isnull().sum())
```

```python
# Runtime has very few missing values, so we can fill it with the median.
df.loc[:, "runtime"] = df["runtime"].fillna(df["runtime"].median())

# Release date has very few missing values, so we remove those rows.
df = df.dropna(subset=["release_date", "director"]).copy()

print("Missing values after cleaning:")
print(df.isnull().sum())
```

```python
print("Rows before removing zero budget/revenue:", df.shape[0])

df = df[(df["budget"] > 0) & (df["revenue"] > 0)].copy()

print("Rows after removing zero budget/revenue:", df.shape[0])
```

```python
# Convert release_date to datetime safely
df["release_date"] = pd.to_datetime(df["release_date"], errors="coerce")

# Remove rows where release_date could not be converted
df = df.dropna(subset=["release_date"]).copy()

# Create release year
df["release_year"] = pd.DatetimeIndex(df["release_date"]).year

display(df[["title", "release_date", "release_year"]].head())
```

```python
df.loc[:, "success"] = df["revenue"] > df["budget"]

display(df[["title", "budget", "revenue", "success"]].head())
```

```python
df.loc[:, "profit"] = df["revenue"] - df["budget"]
df.loc[:, "roi"] = df["profit"] / df["budget"]

df.loc[:, "main_genre"] = df["genres"].apply(lambda x: x[0] if len(x) > 0 else "Unknown")
df.loc[:, "main_actor"] = df["cast"].apply(lambda x: x[0] if len(x) > 0 else "Unknown")

display(df[["title", "main_genre", "main_actor", "director", "profit", "roi"]].head())
```

```python
print("Final cleaned dataset shape:", df.shape)
print("\nFinal columns:")
print(df.columns.tolist())

print("\nData types:")
print(df.dtypes)

display(df.head())
```

```python
df.to_csv("../Datasets/clean_movies.csv", index=False)

print("Saved clean_movies.csv successfully!")

print("\nFiles in current folder:")
print(os.listdir())
```
